```python
import pandas as pd
from langdetect import detect
import sys
sys.path.append("../")
from functions import preprocess_desc_for_classif, extract_skills_job, classify_job, predict_salaire_dh, extract_info_job
```

```python
rekrute = pd.read_csv("rekrute.csv")
emploi_ma = pd.read_csv("emploi.ma.csv")
khdma_ma = pd.read_csv("khdma.ma.csv")
marocannonces = pd.read_csv("maroc_annonces.csv")
```

```python
khdma_ma["pays"] = khdma_ma["pays"].replace("MA", "Maroc")
```

```python
rekrute["source"] = "Rekrute"
emploi_ma["source"] = "Emploi.ma"
khdma_ma["source"] = "Khdma.ma"
marocannonces["source"] = "Maroc Annonces"
```

```python
df = pd.concat([rekrute, emploi_ma, khdma_ma, marocannonces], ignore_index=True)
```

```python
df.rename(columns={"titre": "title"}, inplace=True)
df.rename(columns={"date_poste": "date"}, inplace=True)
df.rename(columns={"date_fin": "deadline"}, inplace=True)
df.rename(columns={"ville": "city"}, inplace=True)
df.rename(columns={"pays": "country"}, inplace=True)
df.rename(columns={"entreprise": "company"}, inplace=True)
df.rename(columns={"langue": "lang"}, inplace=True)
```

```python
df["link"] = "#"
df["company_logo"] = "#"
```

```python
df.columns
```

```python
df = df.drop(columns=["industrie", "contrat", "salaire"])
```

```python
df["domain"] = df["description"].apply(lambda x: classify_job(x))
```

```python
df["hskills"] = df["description"].apply(lambda x: ", ".join(extract_skills_job(x)))
```

```python
df["salary_estimation"] = df["description"].apply(lambda x: predict_salaire_dh(x))
```

```python
for i in range(len(df)):
    infos = extract_info_job(df["description"][i])
    df["soft_skills"][i] = ", ".join(list(infos["SOFT-SKILL"]))
    df["profile"][i] = ", ".join(list(infos["POSTE"]))
    df["education"][i] = ", ".join(list(infos["EDUCATION"]))
    df["experience"][i] = ", ".join(list(infos["EXPERIENCE"]))
    df["langues"][i] = ", ".join(list(infos["LANGUE"]))
```

```python
df.to_csv("static_data.csv", index=False)
```
